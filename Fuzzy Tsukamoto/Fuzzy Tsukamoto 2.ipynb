```python
def kualitas_pelayanan(x):
    if (x <= 3 and x >= 0):
        nilai = 'buruk'
    elif (x > 3 and x <= 8):
        nilai = 'baik'
    elif (x > 8 and x <= 10):
        nilai = 'sangat baik'
    else:
        print("Gaada lah woi")
    return nilai

def kualitas_makanan(x):
    if (x <= 3 and x >= 0):
        nilai = 'buruk'
    elif (x > 3 and x <= 8):
        nilai = 'layak'
    elif (x > 8 and x <= 10):
        nilai = 'sangat enak'
    else:
        print("Gaada lah woi")
    return nilai

def pemberian_tip(x):
    if (x <= 3 and x >= 0):
        nilai = 'rendah'
    elif (x > 3 and x <= 8):
        nilai = 'sedang'
    elif (x > 8 and x <= 10):
        nilai = 'tinggi'
    else:
        print("Gaada lah woi")
    return nilai

def rule1(x1, x2):
    if (kualitas_pelayanan(x1) == 'sangat baik' and kualitas_makanan(x2) == 'sangat enak'):
        return 'tinggi'
    elif(kualitas_pelayanan(x1) == 'baik' and kualitas_makanan(x2) == 'sangat enak'):
         return 'tinggi'
    elif(kualitas_pelayanan(x1) == 'sangat baik' and kualitas_makanan(x2) == 'layak'):
         return 'tinggi'
    else:
        return 'tidak_tinggi'

def rule2(x1, x2):
    if (kualitas_pelayanan(x1) == 'baik' and kualitas_makanan(x2) == 'layak'):
        return 'sedang'
    
    elif(kualitas_pelayanan(x1) == 'sangat baik' and kualitas_makanan(x2) == 'buruk'):
        return 'sedang'
    
    elif(kualitas_pelayanan(x1) == 'buruk' and kualitas_makanan(x2) == 'sangat enak'):
        return 'sedang'
    else:
        return 'tidak_sedang'

def rule3(x1, x2):
    if(kualitas_pelayanan(x1) == 'buruk' and kualitas_makanan(x2) == 'buruk'):
        return 'rendah'
    elif(kualitas_pelayanan(x1) == 'baik' and kualitas_makanan(x2) == 'buruk'):
        return 'sedang'
    
    elif(kualitas_pelayanan(x1) == 'buruk' and kualitas_makanan(x2) == 'layak'):
        return 'sedang'
    else:
        return 'tidak_rendah'
def tsukamoto(x1, x2):
    alpha1 = rule1(x1, x2)
    alpha2 = rule2(x1, x2)
    alpha3 = rule3(x1, x2)

    tip_tinggi = 0.25  
    tip_sedang = 0.15  
    tip_rendah = 0.05  

    if alpha1 == 'tinggi':
        alpha1 = 1.0
    else:
        alpha1 = 0.0

    if alpha2 == 'sedang':
        alpha2 = 1.0
    else:
        alpha2 = 0.0

    if alpha3 == 'rendah':
        alpha3 = 1.0
    else:
        alpha3 = 0.0

    total_tip = (alpha1 * tip_tinggi) + (alpha2 * tip_sedang) + (alpha3 * tip_rendah)

    return total_tip * 100  

nilai_pelayanan = 8
nilai_makanan = 4

tip = tsukamoto(nilai_pelayanan, nilai_makanan)
print("Pemberian tip adalah: {:.2f}%".format(tip))

```

```python
data = [
    (9.8, 6.8),
    (9.5, 9.2),
    (6.4, 9.6),
    (5.7, 6.3),
    (1.3, 2.5),
    (7.4, 6.8)
]

for i, (kualitas_pelayanan_input, kualitas_makanan_input) in enumerate(data, start=1):
    hasil_tip = tsukamoto(kualitas_pelayanan_input, kualitas_makanan_input)
    print(f"Data {i}: Kualitas Pelayanan = {kualitas_pelayanan_input}, Kualitas Makanan = {kualitas_makanan_input}, Prediksi Tip = {hasil_tip:.2f} %")

```

```python
def permintaan_turun(x):
    if (x <= 500):
        nilai = 1
    elif (x >= 1500):
        nilai = 0
    elif (500 <= x <= 1500):
        nilai = (1500 - x)/1000
    return nilai

def permintaan_sedang(x):
    if(x <= 1000):
        nilai = 0
    elif(3500 <= x <= 4500):
        nilai = (4500 - x)/1000
    elif(1000 <= x <= 2000 ):
        nilai = (x - 1000)/1000
    elif(2500 <= x <= 3500):
        nilai = 1
    elif (x >= 4500):
        nilai = 0
    return nilai

    
def permintaan_naik(x):
    if (x >= 5000):
        nilai = 1
    elif (x <= 4000):
        nilai = 0
    elif(4000 <= x <= 5000):
        nilai = (x - 4000)/1000
    return nilai

def persediaan_sedikit(x):
    if(x<=500):
        nilai = 1
    elif(x > 50 and x < 200):
        nilai = (200-x)/(150)
    elif(x >= 200):
        nilai = 0
    return nilai

def persediaan_banyak(x):
    if (x >= 250):
        nilai = 1
    elif (x > 100 and x < 250):
        nilai = (x - 100)/150
    elif (x <= 100):
        nilai = 0
    return nilai

def produksi_sedikit(x):
    if (x <= 1000):
        nilai = 1
    elif (x >= 6000):
        nilai = 0
    elif (x > 1000 and x < 6000):
        nilai = (6000 - x)/5000
    return nilai

def produksi_banyak(x):
    if (x >= 6000):
        nilai = 1
    elif (x <= 1000):
        nilai = 0
    elif (x > 1000 and x < 6000):
        nilai = (x - 1000)/5000
    return nilai


def rule1(x, y):
    if permintaan_turun(x)<=persediaan_banyak(y):
        alpha = permintaan_turun(x)
    else:
        alpha = persediaan_banyak(y)
    z = 6000 - (5000 * alpha)
    return alpha, z;

def rule2(x, y):
    if permintaan_turun(x)<=persediaan_sedikit(y):
        alpha = permintaan_turun(x)
    else:
        alpha = persediaan_sedikit(y)
    z = 6000 - (5000 * alpha)
    return alpha, z;

def rule3(x, y):
    if permintaan_sedang(x)<=persediaan_banyak(y):
        alpha = permintaan_sedang(x)
    else:
        alpha = persediaan_banyak(y)
    z = 1000 + (5000 * alpha)
    return alpha, z;

def rule4(x, y):
    if permintaan_sedang(x)<=persediaan_sedikit(y):
        alpha = permintaan_sedang(x)
    else:
        alpha = persediaan_sedikit(y)
    z = 6000 - (5000 * alpha)
    return alpha, z;

def rule5(x, y):
    if permintaan_naik(x)<=persediaan_banyak(y):
        alpha = permintaan_naik(x)
    else:
        alpha = persediaan_banyak(y)
    z = 1000 + (5000 * alpha)
    return alpha, z;

def rule6(x, y):
    if permintaan_naik(x)<=persediaan_sedikit(y):
        alpha = permintaan_naik(x)
    else:
        alpha = persediaan_sedikit(y)
    z = 1000 + (5000 * alpha)
    return alpha, z;

# Metode Tsukamoto
def tsukamoto(x, y):
    alpha1, z1 = rule1(x, y)
    alpha2, z2 = rule2(x, y)
    alpha3, z3 = rule3(x, y)
    alpha4, z4 = rule4(x, y)
    alpha5, z5 = rule5(x, y)
    alpha6, z6 = rule6(x, y)

    alpha = (alpha1 * z1) + (alpha2 * z2) + (alpha3 * z3) + (alpha4 * z4) + (alpha5 * z5) + (alpha6 * z6) 
    z = alpha1 + alpha2 + alpha3 + alpha4 + alpha5 + alpha6
    return alpha/z

p = 4000
q = 500


print("Jika Permintaan =" ,p, "dan Persediaan = " ,q, "maka Jumlah Sarung yang Harus Diproduksi = ",tsukamoto(p,q) )

```

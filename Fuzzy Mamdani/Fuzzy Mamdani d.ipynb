```python
import numpy as np

def predict_grade(uts_nilai, uas_nilai, tugas_nilai):
    uts_low = np.array([0, 30, 60])
    uts_medium = np.array([30, 70, 100])
    uts_high = np.array([70, 100, 100])
    uas_low = np.array([0, 30, 60])
    uas_medium = np.array([30, 70, 100])
    uas_high = np.array([70, 100, 100])
    tugas_low = np.array([0, 30, 60])
    tugas_medium = np.array([30, 70, 100])
    tugas_high = np.array([70, 100, 100])

    uts_low_membership = np.interp(uts_nilai, uts_low, [0, 1, 0])
    uts_medium_membership = np.interp(uts_nilai, uts_medium, [0, 1, 0])
    uts_high_membership = np.interp(uts_nilai, uts_high, [0, 1, 0])
    uas_low_membership = np.interp(uas_nilai, uas_low, [0, 1, 0])
    uas_medium_membership = np.interp(uas_nilai, uas_medium, [0, 1, 0])
    uas_high_membership = np.interp(uas_nilai, uas_high, [0, 1, 0])
    tugas_low_membership = np.interp(tugas_nilai, tugas_low, [0, 1, 0])
    tugas_medium_membership = np.interp(tugas_nilai, tugas_medium, [0, 1, 0])
    tugas_high_membership = np.interp(tugas_nilai, tugas_high, [0, 1, 0])

    rules = [
        (min(uts_low_membership, uas_low_membership, tugas_low_membership), "grade_low"), #1+1+1
        (min(uts_medium_membership, uas_low_membership, tugas_low_membership), "grade_low"), #2+1+1
        (min(uts_high_membership, uas_low_membership, tugas_low_membership), "grade_medium"), #3+1+1
        (min(uts_low_membership, uas_medium_membership, tugas_low_membership), "grade_low"), #1+2+1
        (min(uts_medium_membership, uas_medium_membership, tugas_low_membership), "grade_medium"), #2+2+1
        (min(uts_high_membership, uas_medium_membership, tugas_low_membership), "grade_medium"), #3+2+1
        (min(uts_low_membership, uas_high_membership, tugas_low_membership), "grade_medium"), #1+3+1
        (min(uts_medium_membership, uas_high_membership, tugas_low_membership), "grade_medium"), #2+3+1
        (min(uts_high_membership, uas_high_membership, tugas_low_membership), "grade_medium"),#3+3+1
        (min(uts_low_membership, uas_low_membership, tugas_medium_membership), "grade_low"), #1+1+2
        (min(uts_medium_membership, uas_low_membership, tugas_medium_membership), "grade_medium"), #2+1+2
        (min(uts_high_membership, uas_low_membership, tugas_medium_membership), "grade_medium"),#3+1+2
        (min(uts_low_membership, uas_low_membership, tugas_high_membership), "grade_medium"), #1+1+3
        (min(uts_medium_membership, uas_low_membership, tugas_high_membership), "grade_medium"), #2+1+3
        (min(uts_high_membership, uas_low_membership, tugas_high_membership), "grade_medium"),#3+1+3
        (min(uts_low_membership, uas_medium_membership, tugas_medium_membership), "grade_medium"), #1+2+2
        (min(uts_medium_membership, uas_medium_membership, tugas_medium_membership), "grade_medium"), #2+2+2
        (min(uts_high_membership, uas_medium_membership, tugas_medium_membership), "grade_medium"),#3+2+2
        (min(uts_low_membership, uas_high_membership, tugas_medium_membership), "grade_medium"), #1+3+2
        (min(uts_medium_membership, uas_high_membership, tugas_medium_membership), "grade_medium"), #2+3+2
        (min(uts_high_membership, uas_high_membership, tugas_medium_membership), "grade_high"),#3+3+2
        (min(uts_low_membership, uas_medium_membership, tugas_high_membership), "grade_medium"), #1+2+3
        (min(uts_medium_membership, uas_medium_membership, tugas_high_membership), "grade_medium"), #2+2+3
        (min(uts_high_membership, uas_medium_membership, tugas_high_membership), "grade_high"),#3+2+3
        (min(uts_low_membership, uas_high_membership, tugas_high_membership), "grade_medium"), #1+3+3
        (min(uts_medium_membership, uas_high_membership, tugas_high_membership), "grade_high"), #2+3+3
        (min(uts_high_membership, uas_high_membership, tugas_high_membership), "grade_high"),#3+3+3
    ]

    grades = {
        "grade_low": 0,
        "grade_medium": 0,
        "grade_high": 0,
    }

    for rule in rules:
        membership, grade = rule
        grades[grade] = max(grades[grade], membership)

    defuzzified_grade = (grades["grade_low"] * 30 + grades["grade_medium"] * 70 + grades["grade_high"] * 100) / (
            grades["grade_low"] + grades["grade_medium"] + grades["grade_high"])

    if defuzzified_grade <= 40:
        output_grade = "Low (C)"
    elif defuzzified_grade <= 70:
        output_grade = "Medium (B)"
    else:
        output_grade = "High (A)"

    return output_grade

num_mhs = int(input("Masukkan jumlah mahasiswa: "))
mhs = []

for i in range(num_mhs):
    nama = input(f"Masukkan nama mahasiswa {i + 1}: ")
    uts = float(input(f"Masukkan nilai UTS untuk mahasiswa {i + 1}: "))
    uas = float(input(f"Masukkan nilai UAS untuk mahasiswa {i + 1}: "))
    tugas_1 = float(input(f"Masukkan nilai Tugas 1 untuk mahasiswa {i + 1}: "))
    tugas_2 = float(input(f"Masukkan nilai Tugas 2 untuk mahasiswa {i + 1}: "))
    tugas_3 = float(input(f"Masukkan nilai Tugas 3 untuk mahasiswa {i + 1}: "))
    rata_rata_tugas = (tugas_1 + tugas_2 + tugas_3) / 3
    grade = predict_grade(uts, uas, rata_rata_tugas)
    mhs.append({"nama": nama, "uts": uts, "uas": uas, "rata_rata_tugas": rata_rata_tugas, "grade": grade})

for i, mhs in enumerate(mhs):
    print(f"Mahasiswa {i + 1}: Nama = {mhs['nama']}, UTS = {mhs['uts']}, UAS = {mhs['uas']}, Rata-rata Tugas = {mhs['rata_rata_tugas']}, Grade = {mhs['grade']}")


```
